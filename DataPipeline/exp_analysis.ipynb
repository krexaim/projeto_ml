```python
from pathlib import Path
import yaml


# Resolve a raiz do projeto subindo a partir do diretório atual até achar o config.
# Funciona independentemente de onde o Jupyter foi iniciado (portável).
def find_project_root(marker="DataPipeline/config.yml"):
    p = Path.cwd().resolve()
    for candidate in [p, *p.parents]:
        if (candidate / marker).exists():
            return candidate
    raise FileNotFoundError("Raiz do projeto não encontrada (DataPipeline/config.yml ausente)")


PROJECT_ROOT = find_project_root()
with open(PROJECT_ROOT / "DataPipeline" / "config.yml", encoding="utf-8") as f:
    cfg = yaml.safe_load(f)

data_dir = PROJECT_ROOT / cfg["paths"]["data_dir"]
target = cfg["project"]["target"]

# EDA é feita sobre o dado LIMPO; a ABT entra apenas para comparar a distribuição do alvo
abt_input_path = data_dir / cfg["data"]["abt_file"]
clean_app_path = data_dir / cfg["data"]["clean_files"]["application"]
```
